```python
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
def relu(x):
  return ((max(0,x)))
```

```python
def plot(x,y,title,grid):
  plt.figure(figsize=(5,5)).gca()
  plt.plot(x,y)
  plt.xlabel('X')
  plt.ylabel('Y')
  plt.grid(grid)
  plt.title(title)
```

##Initialising input points for all activation functions

```python
#Binary Function
def binary(x):
    if(x>=0):
        return 1
    else :
        return 0

x = [float(x)/100 for x in range(-500,500)] 
y = [binary(x2) for x2 in x] 
title = 'Binary Function'
```

```python
#Step Function
a = 4
def step(x):
    if(x>=0):
        return a
    else :
        return 0

x1 = [float(x)/100 for x in range(-500,500)] 
y1 = [step(x) for x in x1]  

title1 = 'Step Function'
```

```python
#Linear Function
m=2
c=0
def linear(x):
    return m*x+c

x2 = [float(x)/100 for x in range(-500,500)] 
y2 = [linear(x1) for x1 in x2] 
title2 = 'Linear Function'
```

```python
#Relu Function
x3 = [float(x)/100 for x in range(-100,100)]
y3 = [relu(x1) for x1 in x3]
title3 = 'ReLu Function'
```

```python
#Leaky relu Function
a = 0.3
def Leakyrelu(x):
    if(x>=0):
        return x
    else:
        return a*x

x4 = [float(x)/100 for x in range(-500,500)] 
y4 = [Leakyrelu(x1) for x1 in x4]  
title4 = 'Leaky Relu Function'
```

```python
#Sigmoid Function
from math import exp
def sigmoid(x):
    return (exp(x)/(1+exp(x)))
x5 = [float(x1)/100 for x1 in range(-500,500)]
y5 = [sigmoid(x2) for x2 in x5]
title5 = 'Sigmoid Function'
```

```python
#Tanh Function
def tanh(x):
    return 2 * sigmoid(2*x) - 1

x6 = [float(x)/100 for x in range(-500,500)] 
y6 = [tanh(x2) for x2 in x6]
title6 = 'Tanh Function'
```

```python
#Swish Function
bta = 1
def swish(x):
    return 2*x*sigmoid(bta*x)
x7 = [float(x)/100 for x in range(-500,500)]
y7 = [swish(x2) for x2 in x7] 
title7 = 'Swish Function'
```

##Plotting The Functions :

```python
main_x = (x,x1,x2,x3,x4,x5,x6,x7)
main_y = (y,y1,y2,y3,y4,y5,y6,y7)
main_title = (title,title1,title2,title3,title4,title5,title6,title7)
grid_main = (True,False,True,False,False,False,False,True)
```

```python
for i in range(0,7):
  plot(x=main_x[i],y=main_y[i],title=main_title[i],grid=grid_main[i])
```
