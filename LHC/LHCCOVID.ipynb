```python
confirmedcases = pd.read_csv("time_series_covid19_confirmed_global.csv")
recoveredcases = pd.read_csv("time_series_covid19_recovered_global.csv")
deaths = pd.read_csv("time_series_covid19_deaths_global.csv")
confirmedcases = confirmedcases.drop(['Province/State','Lat','Long'],axis=1)
recoveredcases = recoveredcases.drop(['Province/State','Lat','Long'],axis=1)
deaths = deaths.drop(['Province/State','Lat','Long'],axis=1)
```

TASK 1

```python
indiaconfirmedcases = confirmedcases.loc[confirmedcases['Country/Region']=='India']
```

```python
indiarecoveredcases = recoveredcases.loc[recoveredcases['Country/Region']=='India']
indiadeathcases = deaths.loc[deaths['Country/Region']=='India']
```

```python
indiaconfirmedcases = pd.DataFrame(indiaconfirmedcases)
indiarecoveredcases = pd.DataFrame(indiarecoveredcases)
indiadeathcases = pd.DataFrame(indiadeathcases)
```

```python
indiaconfirmedcases=indiaconfirmedcases.drop(['Country/Region'],axis=1)
indiarecoveredcases=indiarecoveredcases.drop(['Country/Region'],axis=1)
indiadeathcases = indiadeathcases.drop(['Country/Region'],axis=1)
```

```python
indiaconfirmedcases.index = ["Confirm_Cases"]
indiaconfirmedcases = indiaconfirmedcases.transpose()
```

```python
indiaconfirmedcases
```

```python
indiarecoveredcases.index = ["Recovered_Cases"]
indiarecoveredcases = indiarecoveredcases.transpose()
indiadeathcases.index = ["Deaths"]
indiadeathcases = indiadeathcases.transpose()
```

```python
indiadata = indiaconfirmedcases
```

```python
indiadata['Recovered_Cases'] = indiarecoveredcases
```

```python
indiadata['Deaths'] = indiadeathcases
```

```python
indiadata
```

```python
plt.plot(indiadata.index, indiadata['Confirm_Cases'], label = "Confirmed Cases", linestyle="-")
plt.plot(indiadata.index, indiadata['Recovered_Cases'], label = "Recovered Cases", linestyle="--")
plt.plot(indiadata.index, indiadata['Deaths'], label = "Deaths", linestyle="-.")
plt.xlabel("Days")
plt.ylabel("Number of Cases")
plt.title("Indian Covid Cases")
plt.legend()
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
```

```python
X_features = np.array(range(1,469))
X_features = X_features.reshape(-1,1)
Y = np.array(indiaconfirmedcases['Confirm_Cases'])
Y = Y.reshape(-1,1)
```

```python
x= np.array([469]).reshape(-1,1)
```

```python
from sklearn.preprocessing import PolynomialFeatures
poly = PolynomialFeatures(degree = 10)
X_poly = poly.fit_transform(X_features)
poly = poly.fit(X_poly, Y)
lin2 = LinearRegression()
lin2.fit(X_poly, Y)
ans2 =int(lin2.predict(poly.fit_transform(x))) 
print(" the number of covid cases on 4th May, 2021:",ans2)
```

```python
casetoday=ans2- Y[-1]
```

```python
casetoday
```

```python
import os
import glob
location="LHC"

file_name=[]
for file in os.listdir(location):
        file_name.append(file)
      
print(file_name)
combinedCSV=pd.concat([pd.read_csv("LHC/"+f) for f in file_name])
combinedCSV.to_csv("combineddata.csv",index=False,encoding='utf-8-sig')
```

TASK 5

```python
from sklearn.linear_model import LinearRegression
X_Gujfeatures = np.array(range(1,65))
X_Gujfeatures = X_Gujfeatures.reshape(-1,1)
YGuj = np.array(gujaratdata['Confirmed'])
YGuj = Y_Guj.reshape(-1,1)
X_Gujfeatures.size
```

```python

from sklearn.preprocessing import PolynomialFeatures
poly = PolynomialFeatures(degree = 5)
X_poly = poly.fit_transform(X_Gujfeatures)
poly = poly.fit(X_poly, YGuj)
lin2 = LinearRegression()
lin2.fit(X_poly, YGuj)
x = np.array([65]).reshape(-1, 1)
ans2 =int(lin2.predict(poly.fit_transform(x))) 
print(ans2)
```

```python
casesgujtoday=ans2-YGuj[-1]
casesgujtoday
```

Task 6

```python

print("Cases all over India on 4th May,2021 : ",casetoday)
print("Cases in Gujarat on 4th May,2021 : ",casesgujtoday)
perct = (casesgujtoday[0]/(casetoday[0]+casesgujtoday[0]))*100

plt.pie([perct.round(2),100-perct.round(2)],colors=['Blue','red'],labels=['Gujarat','India'],
        explode=(0.2,0),shadow=True,autopct='%1.1f%%')
plt.title("Covid cases on 4th May")
plt.legend()
plt.show()

```
